## Color Transparency
**RGBA** - Red, Green, Blue, Alpha (0 - transparent)

```python
red = Image.open('images/red_color.jpg')
```

```python
red
```

```python
blue = Image.open('images/blue_color.png')
```

```python
blue = blue.convert('RGB')
```

```python
blue.putalpha(128)
```

```python
blue
```

```python
red.putalpha(128)
```

```python
red
```

```python
blue.paste(red, box=(0, 0), mask=red)
```

```python
blue
```

```python
blue.save('images/purple2.png')
```

```python
purple = Image.open('images/purple2.png')
```
